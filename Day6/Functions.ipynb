```python


a = int(input("Enter the first number : "))
b = int(input("Enter the second number : "))
op = input("Enter the operation : ")
def Calc(a,b,operator):
    if(operator == "+"):
        print(a+b)
    elif(operator == "-"):
        print(a-b)
    elif(operator == "*"):
        print(a*b)
    elif(operator == "/"):
        print(a//b)
    elif(operator == "sq"):
        print(a*a)
    elif(operator == "cb"):
        print(a*a*a)
    elif(operator == "//"):
        print(a//b)
    elif(operator == "%"):
        print(a%b)
    else:
        print("Invalid operator")

Calc(a,b,op)
    
```
