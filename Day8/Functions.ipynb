```python
#Global variable
city = "Bokaro"

def display():
    print("Inside Function:", city)

#Function call
display()

#Outside function
print("Outside Function:", city)
```

```python
#Create global variable balance
balance = 1000

def deposit():
    global balance
    balance = balance + 500
    print("Balance after deposit:", balance)

def withdraw():
    global balance
    balance = balance - 200
    print("Balance after withdraw:", balance)

deposit()
withdraw()
```
