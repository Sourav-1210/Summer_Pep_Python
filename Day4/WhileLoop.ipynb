```python
#Multiplication table of a number
num = int(input("Enter the number :"))
i = 1
while(i<11):
    print(num ,"*",i," = ",num*i)
    i+=1
```

```python
#Count number of digits in a number
num = int(input("Enter the number "))

count = 0
while(num>0):
    count+=1
    num//=10

print(count)
```

```python
#Reverse a number
num = int(input("Enter the number : "))
rev = 0
temp = num
while(temp>0):
    digit = temp%10
    rev = rev*10+digit
    temp//=10

print(rev)
```

```python
#Check whether number is palindrome or not
num = int(input("Enter the number : "))
rev = 0
temp = num
while(temp>0):
    digit = temp%10
    rev = rev*10+digit
    temp//=10

if(rev==num):
    print("It is a palindrome")
else:
    print("It is not a palindrome")
```

```python
#Print sum of all digits of a number
num = int(input("Enter the number : "))
sum = 0
while(num>0):
    digit = num%10
    sum+=digit
    num//=10

print(sum)
```

```python
#Print numbers from 1 to 20
#Skip all multiples of 3
i = 1
while(i<21):
    if(i%3==0):
        i+=1
        continue

    print(i)
    i+=1
```

```python
#Print numbers from 1 to 50
#Skip all the multiples of 4

i = 1
while(i<51):
    if(i%4==0):
        i+=1
        continue
    print(i)
    i+=1
```
