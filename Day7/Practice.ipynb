```python
#Area of Rectangle
length = int(input("Enter the length of table : "))
breadth = int(input("Enter the breadth of table : "))

def AreaOfRectangle(length,breadth):
    return length*breadth

area = AreaOfRectangle(length,breadth)
print("Area of Rectangle is ",area)
```

```python
#Function for calculating Simple Interest
principal = int(input("Enter the Principal amount : "))
rate = int(input("Enter the rate of interest : "))
duration = int(input("Enter the duration : "))

def SimpleInterest(principal,rate,duration):
    return (principal*rate*duration)/100

Interest = SimpleInterest(principal,rate,duration)
print("Simple Interest for customers is ",Interest)
```

```python
#Fuction to convert celsius to fahrenheit
temperature = float(input("Enter the temperature i  Celcius : "))

def CelciusToFahrenheit(temperature):
    return (temperature*9/5)+32

Fahrenheit = CelciusToFahrenheit(temperature)
print("Temperature in Fahrenheit : ",Fahrenheit)
```

```python
#Function to calculate final bill amount after adding GST
BillAmount = float(input("Enter the bill amount : "))
gst = float(input("Enter the GST percentage : "))

def CalculateFinalBill(bill_amount,gst):
    gst_amount = (bill_amount*gst)/100
    final_bill = bill_amount+gst_amount

    return final_bill

print("Final Bill Amount : ",CalculateFinalBill(BillAmount,gst))
```

```python
# Details of the family member 
Name = input("Enter the name of member : ")
Age = int(input("Enter the age of member : "))
City = input("Enter the city of member : ")
Occupation = input("Enter the occupation of the memeber : ")
Phone_no = input("Enter the phone_no of member : ")

def family_member(**details):
    print("Family member details : ")
    for key, value in details.items():
        print(key, ": ",value)

family_member(Name=Name,Age=Age,City=City,Occupation=Occupation,Phone_no=Phone_no)
```

```python
#Customer Bill
def final_bill(price,quantity,discount=0):
    total = price*quantity
    bill = total-discount

    print("Total amount : ",total)
    print("Discount : ",discount)
    print("Finall payable amount : ",bill)

price = float(input("Enter price of one item: "))
quantity = int(input("Enter quantity: "))
discount = float(input("Enter discount : "))

final_bill(price,quantity,discount)
```

```python
#Student report
def PassOrFail(name,Laravel,Combinatorial,Ethics,UIUX,Analytical):
    total = Laravel+Combinatorial+Ethics+UIUX+Analytical
    avg = total/5

    if(avg>40):
        print(name," is passed the examination")
    else:
        print(name," is failed the examination")

name = input("Enter the name of student : ")
laravel = float(input("Enter the marks of laravel : "))
combinatorail = float(input("Enter the marks of combinatorial : "))
ethics = float(input("Enter the marks of ethics : "))
Uiux = float(input("Enter the marks of ui/ux : "))
analytical = float(input("Enter the marks of analytical : "))

PassOrFail(name,laravel,combinatorail,ethics,Uiux,analytical)
```

```python
#Employees Salary
def calculateSalary(basic_salary,hra,da):
    final_salary = basic_salary+hra+da
    return final_salary

basic_salary = float(input("Enter the basic salary : "))
hra = float(input("Enter HRA: "))
da = float(input("Enter DA: "))

salary = calculateSalary(basic_salary,hra,da)
print("Final Salary is ",salary)
```
