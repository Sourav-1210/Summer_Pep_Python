```python
building = int(input("Enter the building no. : "))
floor = int(input("Enter the floor no. : "))
room = int(input("Enter the room no. : "))

for i in range(1,building+1):
    print("Building ",i)
    for j in range(1,floor+1):
        print("   Floor ",j)
        for k in range(1,room+1):
            print("  Room ",k)
```
