```python
#Print number from 1 to num
num = int(input("Enter the number :"))
for i in range(1,num+1):
    print(i)
```

```python
#Print even numbers from 1 to num
num = int(input("Enter the number :"))
for i in range(2,num+1,2):
    print(i)
```

```python
#print Odd numbers 1 to num
num = int(input("Enter the number :"))
for i in range(1,num+1,2):
    print(i)
```

```python
#Reverse counting from num to 1
num = int(input("Enter the number :"))
for i in range(num,0,-1):
    print(i)
```

```python
#Multiplication table
num = int(input("Enter the number :"))
for i in range(1,11):
    print(num ,"*",i, "=",num*i)
```

```python
#Sum of first n numbers
n = int(input("Enter the number :"))
sum = 0
for i in range(1,n+1):
    sum+=i
print(sum)
```

```python
#Factorial
num = int(input("Enter the number"))
fact = 1
for i in range(1,num+1):
    fact*=i
print(fact)
```

```python
#Count number of digits in a number
num = int(input("Enter the number :"))
count = 0
for i in str(num):
    count+=1
print(count) 
```

```python
#Sum of digits of a number
num = int(input("Enter the number :"))
sum = 0
for i in str(num):
    sum+=int(i)
print(sum)
```
