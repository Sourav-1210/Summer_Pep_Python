```python
#Check Palindrome number
num = int(input("Enter the number : "))
rev = 0
temp = num
while temp>0:
    digit = temp%10
    rev = rev*10+digit
    temp//=10
if(num == rev):
    print("It is a palindrome")
else:
    print("Not a palindrome")
```

```python
#Reverse a number
num = int(input("Enter a number :"))
rev = 0
while num>0:
    digit = num%10
    rev = rev*10+digit
    num//=10

print(rev)
```

```python
#Prime number
num = int(input("Enter the number : "))
prime = True

if(num<=1):
    prime = False
for i in range(2,num):
    if(num%i==0):
        prime = False
        break

if(prime):
    print("Prime Number")
else:
    print("Not a prime")
```

```python
#Fibbonaci number
num = int(input("Enter the number : "))
a = 0
b = 1
for i in range(1,num):
    c = a+b
    a = b
    b = c
print(b)
```
